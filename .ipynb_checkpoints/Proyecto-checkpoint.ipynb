```python
fecha = np.arange('2020-11-01', '2021-04-30', dtype='datetime64[D]')

plt.figure()
# Graficar 
plt.plot(fecha, 'ro-.', fecha, 'yo-.')
ax = plt.gca() # proyeccion actual
for item in ax.xaxis.get_ticklabels():
       item.set_rotation(0) 
# Eje X
ax.set_xlabel('cantidad')
# Eje Y
ax.set_ylabel('fecha')
ax.set_title('POS vs PRE') # note la expresión matemática
plt.show()
plt.close()
```
